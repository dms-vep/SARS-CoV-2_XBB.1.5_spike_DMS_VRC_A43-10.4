# Make logoplots showing key sites of escape

```python
import functools
import itertools
import operator
import os

import altair as alt

import dmslogo
import dmslogo.colorschemes

import matplotlib
import matplotlib.pyplot as plt

import matplotlib
import matplotlib.pyplot as plt

import pandas as pd
import yaml

_ = alt.data_transformers.disable_max_rows()
plt.rcParams['svg.fonttype'] = 'none'
```

```python
min_entry = -3
min_ACE2 = -3
n_sites_to_show = 10
```

```python
per_antibody_escape = pd.read_csv(snakemake.input.per_antibody_escape)
# per_antibody_escape = pd.read_csv('../results/summaries/antibody_escape.csv')
```

```python
per_antibody_escape = per_antibody_escape.dropna()

filtered_df = per_antibody_escape[
    (per_antibody_escape['entry in 293T cells'] >= min_entry) &
    (per_antibody_escape['ACE2 binding'] >= min_ACE2)
]
```

```python
# for coloring by ACE2
ace2_colormap = dmslogo.colorschemes.ValueToColorMap(
    minvalue=max(-2, filtered_df["ACE2 binding"].min()),
    maxvalue=0,
    cmap="Blues",
)

filtered_df["color"] = (
    filtered_df["ACE2 binding"]
    .clip(lower=ace2_colormap.minvalue, upper=ace2_colormap.maxvalue)
    .map(ace2_colormap.val_to_color)
)

for orientation in ["horizontal", "vertical"]:
    fig, _ = ace2_colormap.scale_bar(
        orientation=orientation, label="ACE2 binding",
    )
    display(fig)
    svg = os.path.join("results/escape_logos",f"key_escape_sites_ace2_scalebar_{orientation}.svg")
    print(f"Saving to {svg}")
    fig.savefig(svg, bbox_inches="tight")
    plt.close(fig)
```

```python
filtered_df['site_escape'] = filtered_df.groupby('site')['mAb-A43-10-4'].transform('sum')
filtered_df.head()
```

```python
# Identify the top 10 sites by site_escape
top10_sites = (
    filtered_df[['site', 'site_escape']]
    .drop_duplicates(subset='site')
    .sort_values(by='site_escape', ascending=False)
    .head(10)['site']
)

# Filter the original filtered_df to only include those top 10 sites
filtered_df_top10 = filtered_df[filtered_df['site'].isin(top10_sites)]

filtered_df_top10 = filtered_df_top10.rename(columns={'mAb-A43-10-4': 'antibody_escape'})
```

```python
fig, ax = dmslogo.draw_logo(
    data=filtered_df_top10,
    x_col="site",
    letter_col="mutant",
    letter_height_col="antibody_escape",
    color_col="color",
    xtick_col="site"
)
display(fig)
svg = os.path.join("results/escape_logos",f"escape_logos.svg")
print(f"Saving to {svg}")
fig.savefig(svg, bbox_inches="tight")
plt.close(fig)
```
